```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results_df = pd.read_csv("results_total.csv")
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.title("Box loss")
plt.plot(results_df["epoch"], results_df["train/box_loss"], label="train")
plt.plot(results_df["epoch"], results_df["val/box_loss"], label="val")
plt.legend()

plt.subplot(132)
plt.title("Cls loss")
plt.plot(results_df["epoch"], results_df["train/cls_loss"], label="train")
plt.plot(results_df["epoch"], results_df["val/cls_loss"], label="val")
plt.legend()

# plt.subplot(133)
# plt.plot(results_df["epoch"], results_df["train/dfl_loss"], label="train")
# plt.plot(results_df["epoch"], results_df["val/dfl_loss"], label="val")
plt.legend()
```
